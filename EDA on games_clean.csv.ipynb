```python
# Create a helper function to convert K values to numbers

def convert_k_to_numeric(value):
    if pd.isna(value):
        return value
    value = str(value).strip()
    if value.endswith('K'):
        return float(value.replace('K', '')) * 1000
    else:
        return float(value)
```

```python
df1['Times Listed'] = df1['Times Listed'].apply(convert_k_to_numeric)
df1['Number of Reviews'] = df1['Number of Reviews'].apply(convert_k_to_numeric)
df1['Plays'] = df1['Plays'].apply(convert_k_to_numeric)
df1['Playing'] = df1['Playing'].apply(convert_k_to_numeric)
df1['Backlogs'] = df1['Backlogs'].apply(convert_k_to_numeric)
df1['Wishlist'] = df1['Wishlist'].apply(convert_k_to_numeric)

# printing converted columns
print(df1['Times Listed'].head(2))
print(df1['Number of Reviews'].head(2))
print(df1['Plays'].head(2))
print(df1['Playing'].head(2))
print(df1['Backlogs'].head(2))
print(df1['Wishlist'].head(2))
```

## EDA Questions for games_cleaned.csv

#### Question 1: Top-rated games by user reviews


```python
# Top 10 games by Rating
top_rated = df[['Title', 'Rating']].sort_values(by='Rating', ascending=False).head(10)
print("Top 10 Rated Games:\n", top_rated)
```

#### Question 2: Top 10 most wishlisted games

```python
# Top 10 games by Wishlist
top_wishlist = df[['Title', 'Wishlist']].sort_values(by='Wishlist', ascending=False).head(10)
print("Top 10 Most Wishlisted Games:\n", top_wishlist)
```

#### Question 3: Most played games

```python
# Top 10 games by Plays
top_played = df[['Title', 'Plays']].sort_values(by='Plays', ascending=False).head(10)
print("Top 10 Most Played Games:\n", top_played)
```

#### Question 4: Games with highest backlog compared to wishlist

```python
# Create a new column: backlog to wishlist ratio
df['Backlog_to_Wishlist_Ratio'] = df['Backlogs'] / df['Wishlist']

# Top 10 games with highest backlog-to-wishlist ratio
high_backlog_ratio = df[['Title', 'Backlogs', 'Wishlist', 'Backlog_to_Wishlist_Ratio']] \
    .sort_values(by='Backlog_to_Wishlist_Ratio', ascending=False).head(10)

print("Top 10 Games with Highest Backlog vs Wishlist Ratio:\n", high_backlog_ratio)
```

#### Question 5: Distribution of user ratings

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ratings distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Game Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Games')
plt.show()
```

#### Question 6: Average number of plays per genre

```python
# Average plays per genre
avg_plays_genre = df.groupby('Genres')['Plays'].mean().sort_values(ascending=False)
print("Average Plays per Genre:\n", avg_plays_genre.head(10))
```

#### Question 7: Most common genres in the dataset

```python
# Count of games per genre
genre_counts = df['Genres'].value_counts()
print("Most Common Genres:\n", genre_counts.head(10))
```

#### Question 8: Top developers (Teams) by average rating

```python
# Average rating per Team
avg_rating_team = df.groupby('Team')['Rating'].mean().sort_values(ascending=False)
print("Top 10 Developers by Average Rating:\n", avg_rating_team.head(10))
```

#### Question 9: Developer productivity (number of games per team)

```python
# Count of games per Team
games_per_team = df['Team'].value_counts()
print("Top 10 Most Productive Developers (Number of Games):\n", games_per_team.head(10))
```

#### Question 10: Release trend over years

```python
# Extract year from Release Date
df['Release_Year'] = df['Release Date'].dt.year

# Count of games released per year
release_trend = df['Release_Year'].value_counts().sort_index()

# Plot the trend
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
release_trend.plot(kind='line', marker='o', color='green')
plt.title('Number of Games Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()
```

#### 11. Plays vs Ratings (scatter plot)

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Plays', y='Rating')
plt.title('Plays vs Rating')
plt.xlabel('Number of Plays')
plt.ylabel('Rating')
plt.show()
```

#### 12. Wishlist vs Backlogs (scatter plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Wishlist', y='Backlogs')
plt.title('Wishlist vs Backlogs')
plt.xlabel('Wishlist')
plt.ylabel('Backlogs')
plt.show()
```

#### 13. Average Plays per Developer (Team)

```python
avg_plays_team = df.groupby('Team')['Plays'].mean().sort_values(ascending=False).head(10)
print("Top 10 Teams by Average Plays:\n", avg_plays_team)
```

#### 14. Heatmap of Engagement (Plays, Wishlist, Backlogs)

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Plays', 'Wishlist', 'Backlogs']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Engagement Metrics')
plt.show()
```
