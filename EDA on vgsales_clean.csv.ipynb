## EDA Questions for vgsales_cleaned.csv

#### Question 1: Which region generates the most game sales?

```python
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
print("Region with highest sales: ")
region_sales


# North America contributes the highest share of global video game sales.
```

#### Question 2:  What are the best-selling platforms?

```python
platform_sales = (
    df.groupby('Platform')['Global_Sales']
      .sum()
      .sort_values(ascending=False)
)

print("Best-Selling plateforms are shown below:")
platform_sales
```

#### Question 3: What’s the trend of game releases and sales over years?

```python
yearly_trend = (
    df.groupby('Year')
      .agg(
          Total_Games=('Name', 'count'),
          Total_Sales=('Global_Sales', 'sum')
      )
      .sort_index()
)

print("Trend of game releases and sales over years:")
yearly_trend
```

#### Question 4: Who are the top publishers by sales?

```python
publisher_sales = (
    df.groupby('Publisher')['Global_Sales']
      .sum()
      .sort_values(ascending=False)
)

print("Top Publishers by sales: ")
publisher_sales.head(10)
```

#### Question 5: Which games are the top 10 best-sellers globally?

```python
top_10_games = (
    df.sort_values('Global_Sales', ascending=False)
      .head(10)
)

print("Top 10 best sellers games globally: ")

top_10_games[['Name', 'Platform', 'Year', 'Global_Sales']]
```

#### Question 6: How do regional sales compare for specific platforms?

```python
platform_region_sales = (
    df.groupby('Platform')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
      .sum()
      .sort_values(by='NA_Sales', ascending=False)
)

print("Regional sales compare for specific platforms:")
platform_region_sales.head(10)
```

#### Question 7: How has the market evolved by platform over time?

```python
platform_yearly_sales = (
    df.groupby(['Year', 'Platform'])['Global_Sales']
      .sum()
      .reset_index()
      .sort_values(['Year', 'Global_Sales'], ascending=[True, False])
)

print("Market evolved by plateform over time:")
platform_yearly_sales.head(10)
```

#### Question 8:  What are the regional genre preferences?

```python
regional_genre_sales = (
    df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
      .sum()
      .sort_values(by='NA_Sales', ascending=False)
)

print("Regional genre preference shown below:")
regional_genre_sales
```

#### Question 9: What’s the yearly sales change per region?

```python
yearly_region_sales = (
    df.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
      .sum()
      .sort_index()
)

print("Yearly sales change per region:")

yearly_region_sales
```

#### Question 10: What is the average sales per publisher?

```python
avg_sales_publisher = (
    df.groupby('Publisher')['Global_Sales']
      .mean()
      .sort_values(ascending=False)
)

print("Average sales per publisher: ")
avg_sales_publisher.head(10)
```

#### Question 11:  What are the top 5 best-selling games per platform?

```python
top5_per_platform = (
    df.groupby('Platform')
      .apply(lambda x: x.sort_values('Global_Sales', ascending=False).head(5))
      .reset_index(drop=True)
)

print("Top 5 best selling games per plateform:")
top5_per_platform[['Platform', 'Name', 'Year', 'Global_Sales']]
```
