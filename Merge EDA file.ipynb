## Rename merge key columns

```python
# Rename title columns to a common name

df1.rename(columns = {'Title':'Game'}, inplace = True)
df2.rename(columns = {'Name': 'Game'}, inplace = True)
```

```python
# Verifying the change

df1.columns.tolist()
```

```python
df2.columns.tolist()
```

## Normalize game names

```python
# Normalize game names for better matching
df1['Game'] = df1['Game'].str.lower().str.strip()
df2['Game'] = df2['Game'].str.lower().str.strip()
```

```python
# Quick check

print(df1['Game'].head())
```

```python
print(df2['Game'].head())
```

# Important Step: Merge the datasets

```python
# Merge datasets (left join)

merged_df = pd.merge(
    df1,
    df2,
    on ='Game',
    how ='left'
)
```

## Important Step: EDA on merged_games_sales.csv.

### Q1: Which game genres generate the most global sales?

```python
# Filter rows with sales data

sales_data = merged_df[merged_df['Global_Sales'].notna()]
sales_data
```

```python
# Aggregate sales by genre

# Group by Genre (sales-side Genre column)
genre_sales = sales_data.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
genre_sales
```

```python
# View top genres

print(genre_sales.head(10))
```

```python
# Visualize

import matplotlib.pyplot as plt

genre_sales.head(10).plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title("Top 10 Game Genres by Global Sales")
plt.ylabel("Global Sales (Millions)")
plt.xlabel("Genre")
plt.show()
```

### Q2 : How does user rating affect global sales?

```python
#  Filter rows with sales & rating

rating_sales_data = merged_df[merged_df['Global_Sales'].notna() & merged_df['Rating'].notna()]
rating_sales_data
```

```python
# Simple correlation

correlation = rating_sales_data['Rating'].corr(rating_sales_data['Global_Sales'])
print(f"Correlation between Rating and Global Sales: {correlation:.2f}")
```

```python
# Visualize

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.scatterplot(data=rating_sales_data, x='Rating', y='Global_Sales')
sns.regplot(data=rating_sales_data, x='Rating', y='Global_Sales', scatter=False, color='red')
plt.title("User Rating vs Global Sales")
plt.xlabel("Rating")
plt.ylabel("Global Sales (Millions)")
plt.show()
```

### Q3: Which platforms have the most high-rated games?

```python
# Filter high-rated games

high_rated = merged_df[merged_df['Rating'] >= 4]
high_rated
```

```python
# Count high-rated games by platform

platform_counts = high_rated['Platform'].value_counts()
print(platform_counts.head(10))
```

```python
# Visualize

import matplotlib.pyplot as plt

platform_counts.head(10).plot(kind='bar', figsize=(10,5), color='lightgreen')
plt.title("Top Platforms with High-Rated Games (Rating ≥ 4)")
plt.ylabel("Number of Games")
plt.xlabel("Platform")
plt.show()
```

### Q4: What's the trend of releases and sales over time

```python
# Some datasets have 'Release Date' as object, vgsales has 'Year'
# Created a 'Year' column from games dataset
merged_df['Year'] = pd.to_datetime(merged_df['Release Date'], errors='coerce').dt.year
merged_df['Year'] 
```

```python
# Filter rows with sales and valid years

year_data = merged_df[merged_df['Global_Sales'].notna() & merged_df['Year'].notna()]
year_data 
```

```python
# Aggregate by year

yearly_sales = year_data.groupby('Year')['Global_Sales'].sum()
print(yearly_sales)
```

```python
yearly_releases = year_data.groupby('Year').size()
print(yearly_releases)
```

```python
# Visualize

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(yearly_sales.index, yearly_sales.values, label='Global Sales (Millions)', color='blue')
plt.bar(yearly_releases.index, yearly_releases.values, alpha=0.4, label='Number of Releases', color='orange')

plt.title("Game Releases and Global Sales Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Sales / Number of Releases")
plt.legend()
plt.show()
```

### Q5: Do highly wishlisted games lead to more sales?

```python
# Filter rows with Wishlist and Global_Sales

wishlist_data = merged_df[merged_df['Global_Sales'].notna() & merged_df['Wishlist'].notna()]
wishlist_data 
```

```python
# Simple correlation

correlation = wishlist_data['Wishlist'].corr(wishlist_data['Global_Sales'])
print(f"Correlation between Wishlist and Global Sales: {correlation:.2f}")
```

### Q6: Which genres have the highest engagement but lowest sales?

```python
# Define engagement

merged_df['Engagement'] = merged_df['Plays'] + merged_df['Playing'] + merged_df['Backlogs'] + merged_df['Wishlist']
merged_df['Engagement']
```

```python
# Filter rows with sales info

engagement_sales_data = merged_df[merged_df['Global_Sales'].notna()]
engagement_sales_data
```

```python
# Aggregate by Genre

# Average engagement per genre

# Average sales per genre

genre_stats = engagement_sales_data.groupby('Genre').agg({
    'Engagement': 'mean',
    'Global_Sales': 'mean'
}).sort_values(by=['Engagement'], ascending=False)


genre_stats
```

```python
# Find high engagement but low sales

genre_stats['Engagement_to_Sales'] = genre_stats['Engagement'] / (genre_stats['Global_Sales'] + 0.01)
genre_stats.sort_values(by='Engagement_to_Sales', ascending=False).head(10)
```

### Q7: Do highly listed games (Wishlist / Backlogs) correlate with better ratings?

```python
# Filter rows with Rating and engagement data

engagement_rating_data = merged_df[
    merged_df['Rating'].notna() &
    merged_df['Wishlist'].notna() &
    merged_df['Backlogs'].notna()
]

engagement_rating_data 
```

```python
# Compute correlations

wishlist_corr = engagement_rating_data['Wishlist'].corr(engagement_rating_data['Rating'])
backlogs_corr = engagement_rating_data['Backlogs'].corr(engagement_rating_data['Rating'])

print(f"Correlation between Wishlist and Rating: {wishlist_corr:.2f}")
print(f"Correlation between Backlogs and Rating: {backlogs_corr:.2f}")
```

```python
# Visualize

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.scatterplot(x='Wishlist', y='Rating', data=engagement_rating_data)
sns.regplot(x='Wishlist', y='Rating', data=engagement_rating_data, scatter=False, color='red')
plt.title("Wishlist vs Rating")

plt.subplot(1,2,2)
sns.scatterplot(x='Backlogs', y='Rating', data=engagement_rating_data)
sns.regplot(x='Backlogs', y='Rating', data=engagement_rating_data, scatter=False, color='red')
plt.title("Backlogs vs Rating")

plt.tight_layout()
plt.show()
```

### Q9: What are the Top-performing combinations of Genre + Platform

```python
# Filter rows with sales data

genre_platform_data = merged_df[merged_df['Global_Sales'].notna()]
genre_platform_data
```

```python
# Aggregate global sales by Genre + Platform

top_combinations = genre_platform_data.groupby(['Genres', 'Platform'])['Global_Sales'].sum().sort_values(ascending=False)
print(top_combinations.head(10))
```

```python
# Visualize(Heatmap)

import seaborn as sns
import matplotlib.pyplot as plt

# Pivot table for heatmap
heatmap_data = genre_platform_data.pivot_table(
    index='Genres',
    columns='Platform',
    values='Global_Sales',
    aggfunc='sum',
    fill_value=0
)

plt.figure(figsize=(15,8))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5)
plt.title("Global Sales by Genre + Platform")
plt.ylabel("Genre")
plt.xlabel("Platform")
plt.show()
```

### Q10: Regional sales heatmap by Genre

```python
# Filter rows with sales data

regional_data = merged_df[merged_df['Global_Sales'].notna()]
regional_data
```

```python
# Aggregate sales by Genre and Region

regional_sales = regional_data.groupby('Genres')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
print(regional_sales.head(10))
```

```python
# Visualize(Heatmap)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(regional_sales, annot=True, fmt=".1f", cmap="YlOrRd", linewidths=0.5)
plt.title("Regional Sales by Genre (Millions)")
plt.ylabel("Genre")
plt.xlabel("Region")
plt.show()
```
